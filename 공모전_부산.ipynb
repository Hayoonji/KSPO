```python
import pandas as pd
from sklearn.metrics import accuracy_score
sports = pd.read_csv("busan2.csv")
sports=sports.dropna(axis=0)
print(sports.shape)
sports=sports.astype('int')
sports
```

```python
sports=pd.get_dummies(sports , columns = ['Q11'])
sports=pd.get_dummies(sports , columns = ['Q131'])
sports=pd.get_dummies(sports , columns = ['Q08'])
sports=pd.get_dummies(sports , columns = ['Q07'])
sports
```

```python
features=sports[['AGE','Q131_1','Q131_2','Q131_3','Q131_4','Q131_5',
                 'Q08_1','Q08_2','Q08_3','Q08_4','Q08_5','Q08_6','Q07_0','Q07_1']]
know=sports['Q32_b']
from sklearn.model_selection import train_test_split
train_features, test_features, train_labels, test_labels = train_test_split(features, know)
```

```python
#sports=sports[sports['LOC1']==2]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
train_features = scaler.fit_transform(train_features)
test_features = scaler.transform(test_features)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(train_features, train_labels)
```

```python
print(model.score(train_features, train_labels))
print(model.score(test_features, test_labels))
```

```python
import statsmodels.api as sm
X_train_add_constant=sm.add_constant(train_features)
sm_lr_ac=sm.OLS(train_labels,X_train_add_constant).fit()
sm_lr_ac.summary()
```

```python
logit=sm.Logit(train_labels,X_train_add_constant)
results=logit.fit(method = "newton")
results.summary2()
```
