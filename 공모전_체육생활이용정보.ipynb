```python
df2_ratio = df2.copy()
df2_ratio.columns = ['연도','금상','은상','동상','참가상']
df2_ratio['금상'] = df2['금상 수상 건수'] / (df2['금상 수상 건수']+df2['은상 수상 건수']+df2['동상 수상 건수']+df2['참가상 수상 건수'])
df2_ratio['은상'] = df2['은상 수상 건수'] / (df2['금상 수상 건수']+df2['은상 수상 건수']+df2['동상 수상 건수']+df2['참가상 수상 건수'])
df2_ratio['동상'] = df2['동상 수상 건수'] / (df2['금상 수상 건수']+df2['은상 수상 건수']+df2['동상 수상 건수']+df2['참가상 수상 건수'])
df2_ratio['참가상'] = df2['참가상 수상 건수'] / (df2['금상 수상 건수']+df2['은상 수상 건수']+df2['동상 수상 건수']+df2['참가상 수상 건수'])
df2_ratio
```
