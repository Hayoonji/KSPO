경북 대구 충북 광주 충남 서울 전남 세종 전북 대전 울산 제주 부산 강원 경기 인천 경남

```python
city = '부산광역시'
```

```python
fac_gb = fac.groupby(['시도명','체육시설업종코드명']).시설상태.count().reset_index()
```

```python
fac_gov = fac.loc[fac.시설구분명 == '공공']
fac_gov_gb = fac_gov.groupby(['시도명','체육시설업종코드명']).시설상태.count().reset_index()
```

```python
fac_pri = fac.loc[fac.시설구분명 != '공공']
fac_pri_gb = fac_pri.groupby(['시도명','체육시설업종코드명']).시설상태.count().reset_index()
```

```python
fac_gb['비율'] = fac_gb[fac_gb.시도명 == city].시설상태 / fac_gb[fac_gb.시도명 == city].시설상태.sum()
fac_gov_gb['비율'] = fac_gov_gb[fac_gov_gb.시도명 == city].시설상태 / fac_gov_gb[fac_gov_gb.시도명 == city].시설상태.sum()
fac_pri_gb['비율'] = fac_pri_gb[fac_pri_gb.시도명 == city].시설상태 / fac_pri_gb[fac_pri_gb.시도명 == city].시설상태.sum()
```

## 공공+민간 모두

```python
df2_ratio=df2.copy()
df2_ratio['Gold']=df2_ratio['금상 수상 건수']/df2_ratio['측정횟수']
df2_ratio['Silver']=df2_ratio['은상 수상 건수']/df2_ratio['측정횟수']
df2_ratio['Bronze']=df2_ratio['동상 수상 건수']/df2_ratio['측정횟수']
df2_ratio['Part']=df2_ratio['참가상 수상 건수']/df2_ratio['측정횟수']
df2_gb=df2_ratio.groupby(['센터 시도명'])[['Gold','Silver','Bronze','Part']].mean()*100
df2_gb=df2_gb.reset_index()
df2_gb.sort_values(by='Gold',ascending=False)
df2_ratio=df2.copy()
df2_ratio['튼튼']=(df2_ratio['금상 수상 건수']+df2_ratio['은상 수상 건수']+df2_ratio['동상 수상 건수'])/df2_ratio['측정횟수']
df2_ratio['Part']=df2_ratio['참가상 수상 건수']/df2_ratio['측정횟수']

df2_gb=df2_ratio.groupby(['측정년도','센터 시도명'])[['튼튼','Part']].mean()
df2_gb=df2_gb.reset_index()
df2_gb
```

## 공공만

```python
pub_fac=fac[fac['시설구분명']=='공공']
pub_fac
pub_fac_sum=pub_fac['시도명'].value_counts().sum()
pub_fac_sum
pub_fac_vs=(pub_fac.groupby(['시도명'])['시설명'].count()/pub_fac_sum)*100
pub_fac_vs=pub_fac_vs.to_frame()
pub_fac_vs=pub_fac_vs.reset_index()
pub_fac_vs.sort_values(by='시설명',ascending=False)
merge_pubfac_df2 = pd.merge(pub_fac_vs,df2_gb, how='inner', left_on='시도명', right_on='센터 시도명').drop(columns='센터 시도명')
merge_pubfac_df2.rename(columns={'시설명': '시설개수비율'},inplace=True)
merge_pubfac_df2.sort_values(by='시설개수비율',ascending=False)
merge_pubfac_df2.corr()
```

```python
priv_fac=fac[(fac['시설구분명']=='신고')|(fac['시설구분명']=='등록')]
priv_fac
priv_fac_sum=priv_fac['시도명'].value_counts().sum()
priv_fac_sum
priv_fac_vs=(priv_fac.groupby(['시도명'])['시설명'].count()/priv_fac_sum)*100
priv_fac_vs=priv_fac_vs.to_frame()
priv_fac_vs=priv_fac_vs.reset_index()
priv_fac_vs.sort_values(by='시설명',ascending=False)
merge_privfac_df2 = pd.merge(priv_fac_vs,df2_gb, how='inner', left_on='시도명', right_on='센터 시도명').drop(columns='센터 시도명')
merge_privfac_df2.rename(columns={'시설명': '시설개수비율'},inplace=True)
merge_privfac_df2.sort_values(by='시설개수비율',ascending=False)
merge_privfac_df2.corr()
```
