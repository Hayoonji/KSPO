```python
df2=df[['센터 시도명','시군구','측정년도','측정횟수','유소년 회원 수','청소년 회원 수','성인 회원 수','노인 회원 수','남성 회원수','여성 회원수','금상 수상 건수','은상 수상 건수','동상 수상 건수','참가상 수상 건수']]
df2
```

```python
df2['수상 건수']=df2['금상 수상 건수']+df2['은상 수상 건수']+df2['동상 수상 건수']
df2
```

## 명칭 바꾸기, 통일하기

```python
df2['센터 시도명'].value_counts()
```

```python
df2['센터 시도명']=df2['센터 시도명'].replace('경기','경기도')
df2['센터 시도명']=df2['센터 시도명'].replace('서울','서울특별시')
df2['센터 시도명']=df2['센터 시도명'].replace('충북','충청북도')
df2['센터 시도명']=df2['센터 시도명'].replace('부산','부산광역시')
df2['센터 시도명']=df2['센터 시도명'].replace('전남','전라남도')
df2['센터 시도명']=df2['센터 시도명'].replace('경북','경상북도')
df2['센터 시도명']=df2['센터 시도명'].replace('강원','강원도')
df2['센터 시도명']=df2['센터 시도명'].replace('광주','광주광역시')
df2['센터 시도명']=df2['센터 시도명'].replace('전북','전라북도')
df2['센터 시도명']=df2['센터 시도명'].replace('경남','경상남도')
df2['센터 시도명']=df2['센터 시도명'].replace('인천','인천광역시')
df2['센터 시도명']=df2['센터 시도명'].replace('대전','대전광역시')
df2['센터 시도명']=df2['센터 시도명'].replace('충남','충청남도')
df2['센터 시도명']=df2['센터 시도명'].replace('대구','대구광역시')
df2['센터 시도명']=df2['센터 시도명'].replace('제주','제주특별자치도')
df2['센터 시도명']=df2['센터 시도명'].replace('세종','세종특별자치시')
df2['센터 시도명']=df2['센터 시도명'].replace('울산','울산광역시')
```

## 비율 구하기

```python
df2_ratio=df2.copy()
df2_ratio['Gold']=df2_ratio['금상 수상 건수']/df2_ratio['측정횟수']
df2_ratio['Silver']=df2_ratio['은상 수상 건수']/df2_ratio['측정횟수']
df2_ratio['Bronze']=df2_ratio['동상 수상 건수']/df2_ratio['측정횟수']
df2_ratio['Part']=df2_ratio['참가상 수상 건수']/df2_ratio['측정횟수']
df2_gb=df2_ratio.groupby(['측정년도','센터 시도명'])[['Gold','Silver','Bronze','Part']].mean()*100
```

```python
df2_gb=df2_gb.reset_index()
df2_gb[df2_gb['측정년도']==2021].sort_values(by='Part',ascending=True)
```

```python
year_ratio=df2.copy()
year_ratio['튼튼']=(year_ratio['수상 건수'])/year_ratio['측정횟수']
year_ratio['Part']=year_ratio['참가상 수상 건수']/year_ratio['측정횟수']

year_gb=year_ratio.groupby(['측정년도','센터 시도명'])[['튼튼','Part']].mean()
year_gb=year_gb.reset_index()
year_gb_sort=year_gb[year_gb['측정년도']==2021].sort_values(by='튼튼',ascending=False)
year_gb_sort=year_gb_sort[['측정년도','센터 시도명','튼튼','Part']]

year_gb_sort = year_gb_sort.reset_index()
year_gb_sort = year_gb_sort.drop(columns =['index'])
```
