```python
import pandas as pd
sports = pd.read_csv("OXss.csv")
sports=sports.dropna(axis=0)
sports=sports.astype('int')
sports
```

```python
sports=pd.get_dummies(sports , columns = ['Q14'])
```

```python
sports
```

```python
features=sports[['Q07','SEX','AGE','Q11','Q12','Q051','Q14_b']]
know=sports['Q32_b']

from sklearn.model_selection import train_test_split
train_features, test_features, train_labels, test_labels = train_test_split(features, know)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
train_features = scaler.fit_transform(train_features)
test_features = scaler.transform(test_features)
```

### 로지스틱 회귀분석

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(train_features, train_labels)
```

```python
print(model.score(train_features, train_labels))
print(model.score(test_features, test_labels))
```
